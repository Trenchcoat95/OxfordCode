```python
t0.GetEntry(getentry)

gr_par0_MC = ROOT.TGraph()
for i in range(t0.xyz_plane.size()): gr_par0_MC.SetPoint(i,i,t0.xyz_plane.at(i).Z())


cp0 = ROOT.TCanvas("myCanvasNamep0","The Canvas Titlep0",800,600)
#gr_par0.SetTitle("1/pT evolution : Real Seed, #DeltaE correction;point;(GeV/c)^{-1}")
gr_par0_MC.SetMarkerStyle(31)
#gr_par0.SetMarkerStyle(31)
#gr_par0_forw.SetMarkerStyle(31)
gr_par0_MC.SetMarkerColor(2)
gr_par0_MC.Draw()
cp0.Draw()
```

```python
t0.GetEntry(getentry)

gr_par0 = ROOT.TGraph()
for i in range(t0.parvect_bkw.size()): gr_par0.SetPoint(i,i,t0.parvect_bkw.at(t0.parvect_bkw.size()-i-1)[0])

gr_par0_forw = ROOT.TGraph()
for i in range(t0.parvect.size()): gr_par0_forw.SetPoint(i,i+1,t0.parvect.at(i)[0])

gr_par0_MC = ROOT.TGraph()
for i in range(t0.xyz_plane.size()): gr_par0_MC.SetPoint(i,i,t0.xyz_plane.at(i).Y())


cp0 = ROOT.TCanvas("myCanvasNamep0","The Canvas Titlep0",800,600)
gr_par0.SetTitle("1/pT evolution : Real Seed, #DeltaE correction;point;(GeV/c)^{-1}")
gr_par0_MC.SetMarkerStyle(31)
gr_par0.SetMarkerStyle(31)
gr_par0_forw.SetMarkerStyle(31)
gr_par0_MC.SetMarkerColor(2)
gr_par0.SetMarkerColor(4)
gr_par0_forw.SetMarkerColor(3)
gr_par0_MC.Draw()
gr_par0.Draw("LPSAME")
gr_par0_forw.Draw("LPSAME")
cp0.Modified()
cp0.Update()

legend = ROOT.TLegend(0.1,0.8,0.4,0.9)
#legendtext = "Average y RMS: " + "{:.9f}".format(averageyRMS)
legend.AddEntry(gr_par0_forw,"Kalman Y Forward")
legend.AddEntry(gr_par0,"Kalman Y Backwards")
legend.AddEntry(gr_par0_MC,"MC Y")
legend.Draw()

cp0.Draw()
```

```python
t0.GetEntry(getentry)

gr_par1 = ROOT.TGraph()
for i in range(t0.parvect_bkw.size()): gr_par1.SetPoint(i,i,t0.parvect_bkw.at(t0.parvect_bkw.size()-i-1)[1])

gr_par1_forw = ROOT.TGraph()
for i in range(t0.parvect.size()): gr_par1_forw.SetPoint(i,i+1,t0.parvect.at(i)[1])

gr_par1_MC = ROOT.TGraph()
for i in range(t0.xyz_plane.size()): gr_par1_MC.SetPoint(i,i,t0.xyz_plane.at(i).X())


cp1 = ROOT.TCanvas("myCanvasNamep1","The Canvas Titlep1",800,600)
gr_par1.SetTitle("1/pT evolution : Real Seed, #DeltaE correction;point;(GeV/c)^{-1}")
gr_par1_MC.SetMarkerStyle(31)
gr_par1.SetMarkerStyle(31)
gr_par1_forw.SetMarkerStyle(31)
gr_par1_MC.SetMarkerColor(2)
gr_par1.SetMarkerColor(4)
gr_par1_forw.SetMarkerColor(3)
gr_par1_MC.Draw()
gr_par1.Draw("LPSAME")
gr_par1_forw.Draw("LPSAME")
cp1.Modified()
cp1.Update()

legend = ROOT.TLegend(0.1,0.8,0.4,0.9)
#legendtext = "Average y RMS: " + "{:.9f}".format(averageyRMS)
legend.AddEntry(gr_par1_forw,"Kalman X Forward")
legend.AddEntry(gr_par1,"Kalman X Backwards")
legend.AddEntry(gr_par1_MC,"MC X")
legend.Draw()

cp1.Draw()
```

```python
t0.GetEntry(getentry)

gr_par2 = ROOT.TGraph()
for i in range(t0.parvect_bkw.size()): gr_par2.SetPoint(i,i,t0.parvect_bkw.at(t0.parvect_bkw.size()-i-1)[4])

gr_par2_forw = ROOT.TGraph()
for i in range(t0.parvect.size()): gr_par2_forw.SetPoint(i,i+1,t0.parvect.at(i)[4])

gr_par2_MC = ROOT.TGraph()
for i in range(t0.parvect.size()+1): gr_par2_MC.SetPoint(i,i,t0.invpT_MC)

gr_par2_seed = ROOT.TGraph()
for i in range(t0.parvect.size()+1): gr_par2_seed.SetPoint(i,i,t0.invpT_seed)

gr_par2_MC.GetYaxis().SetRangeUser(t0.invpT_MC-0.2,t0.invpT_MC+0.2)

cp2 = ROOT.TCanvas("myCanvasNamep2","The Canvas Titlep2",800,600)
gr_par2.SetTitle("1/pT evolution : Real Seed, #DeltaE correction;point;(GeV/c)^{-1}")
gr_par2_MC.SetMarkerStyle(31)
gr_par2.SetMarkerStyle(31)
gr_par2_seed.SetMarkerStyle(31)
gr_par2_forw.SetMarkerStyle(31)
gr_par2_MC.SetMarkerColor(2)
gr_par2_seed.SetMarkerColor(6)
gr_par2.SetMarkerColor(4)
gr_par2_forw.SetMarkerColor(3)
gr_par2_MC.Draw()
gr_par2.Draw("LPSAME")
gr_par2_forw.Draw("LPSAME")
gr_par2_seed.Draw("LPSAME")
cp2.Modified()
cp2.Update()

legend = ROOT.TLegend(0.1,0.8,0.4,0.9)
#legendtext = "Average y RMS: " + "{:.9f}".format(averageyRMS)
legend.AddEntry(gr_par2_forw,"Kalman q/pT Forward")
legend.AddEntry(gr_par2,"Kalman q/pT Backwards")
legend.AddEntry(gr_par2_MC,"MC q/pT")
legend.Draw()

cp2.Draw()
```
